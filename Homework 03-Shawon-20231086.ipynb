### Q1. Create a Pie Chart and a barplot for any categorical variable. Compare which plot is better and  why? Use markdown cells to write your explanations.

```python
plt.figure(figsize=(12,6))
# Pie Chart
plt.subplot(1, 2, 1)
plt.pie(trans_type, labels=trans_type.index, autopct='%1.1f%%')
plt.title("VW Cars - Transmission Distribution (Pie Chart)")

# barplot
plt.subplot(1, 2, 2)
trans_type.plot(kind="bar")
plt.xlabel("Transmission Type")
plt.ylabel("Count")
plt.title("VW Cars - Transmission Type Distribution (Barplot)")

plt.tight_layout()
plt.show()
```

To compare the Pie Chart and the barplot, let's analyze their strengths and weaknesses using markdown cells:

Pie Chart:

Strengths:
Provides a visual representation of the distribution of body types in the dataset.
Easily shows the proportion of each body type relative to the whole.
Useful for illustrating the "parts of a whole" concept.
Weaknesses:
May become cluttered if there are too many categories.
It can be challenging to compare the size of the slices accurately.
Difficult to represent precise counts or numerical values.
Barplot:

Strengths:
Displays the count of each body type with clear bars.
Allows for easy comparison of the counts between different body types.
Provides a more precise representation of numerical values.
Weaknesses:
Less intuitive than a Pie Chart when it comes to understanding proportions.
The visual representation is less compact compared to a Pie Chart.
In this case, since we are comparing the distribution of different body types, the barplot is a better choice. It provides a clear and straightforward representation of the counts for each category, allowing for easy comparison. Additionally, the barplot allows for a more accurate understanding of the numerical values.

However, the choice between a Pie Chart and a barplot ultimately depends on the specific data and the purpose of the visualization. Both types of plots have their merits, and the decision should be based on the information being conveyed and the audience's understanding.

### Q3. Create Two regression plots with numeric columns.

```python
# Select the numeric columns for regression plots
mileage = vw_data['mileage']
price = vw_data['price']
mpg=vw_data['mpg']

plt.figure(figsize=(12,6))

# Plot the regression plot of mileage vs. price
plt.subplot(1, 2, 1)
sns.regplot(x=mileage, y=price)
plt.xlabel("mileage")
plt.ylabel("price")
plt.title("VW Cars - mileage vs. price")

# Plot the regression plot of mpg vs. price
plt.subplot(1, 2, 2)
sns.regplot(x=mpg, y=price)
plt.xlabel("mpg")
plt.ylabel("price")
plt.title("VW Cars - mpg vs. price")

plt.tight_layout()
plt.show()
```

```python
# Create a boxplot with a categorical column as x-axis
plt.figure(figsize=(8, 6))
sns.boxplot(data=vw_data, x='fuelType', y='price')
plt.title("VW Cars - Price Distribution by fuelType (Boxplot)")
plt.xlabel("fuelType")
plt.ylabel("price")
plt.show()

# Create a violin plot with a categorical column as hue
plt.figure(figsize=(8, 6))
sns.violinplot(data=vw_data, x='fuelType', y='price', hue='transmission', inner='stick')
plt.title("VW Cars - price Distribution by fuelType (Violinplot)")
plt.xlabel("fuelType")
plt.ylabel("price")
plt.show()

# Create histograms for each numeric column
plt.figure(figsize=(12, 6))
for i, column in enumerate(numeric_columns):
    plt.subplot(1, len(numeric_columns), i + 1)
    sns.histplot(data=vw_data, x=column, kde=True)
    plt.title(f"VW Cars - {column} Distribution")
    plt.xlabel(column)
    plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

### Q10. Perform Exploratory Data Analysis (EDA) using groupby/pivot_table and barplot (total 9 barplots), based on model, transmission, and fuelType, a) What are the top 5 selling car models/transmission/fuelType in the dataset?

```python
grouped = vw_data.groupby(['model', 'transmission', 'fuelType']).size().reset_index(name='count')


sorted_data = grouped.sort_values('count', ascending=False)

top_5_selling_cars = sorted_data.head(5)

print(top_5_selling_cars)
```

### b) What's the average selling price of the top 5 selling car models/transmission/fuelType?

```python
top_5_selling_cars = vw_data['model'].value_counts().head(5).index

filtered_data = vw_data[vw_data['model'].isin(top_5_selling_cars)]

grouped_data = filtered_data.groupby(['model', 'transmission', 'fuelType'])

average_prices = grouped_data['price'].mean()

print(average_prices)
```

### c) What's the total sale of the top 5 selling car models/transmission/fuelType?

```python
top_5_selling_cars = vw_data['model'].value_counts().head(5).index

filtered_data = vw_data[vw_data['model'].isin(top_5_selling_cars)]

grouped_data = filtered_data.groupby(['model', 'transmission', 'fuelType'])

total_sales = grouped_data['price'].sum()

print(total_sales)
```
